```python
# Enter your solution here
df.loc[(df.Group == 'Active') & (df.Age >= 50), ['HR', 'Age']]
```

Display the ages and survey choice of those that did not choose $1$ or $5$ in the `Survey` column.

```python
# Enter your solution here
crit = (df.Survey == 1) | (df.Survey == 5)
df.loc[~crit, ['Age', 'Survey']]
```
